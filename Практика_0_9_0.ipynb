# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def find_unique_elements(list1, list2):
    """Функция нахождения уникальных элементов из двух списков."""
    set1 = set(list1)
    set2 = set(list2)

    unique_from_list1 = set1 - set2
    unique_from_list2 = set2 - set1

    result = unique_from_list1 | unique_from_list2

    return result


# Пример использования.
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
unique_elements = find_unique_elements(list1, list2)
print(unique_elements)
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random


# Состояния питомца.
STATES = {
    'hunger': 100,
    'joy': 100,
    'energy': 100
}

# Имя, тип и возраст питомца.
PET_INFO = {}


def adopt_pet(name, pet_type, age, PET_INFO):
    """Функция для усыновления питомца."""
    PET_INFO['name'] = name
    PET_INFO['type'] = pet_type
    PET_INFO['age'] = age
    print(f"Ваш новый питомец - {name}, он/она - {pet_type}, возрастом {age}.")


def feed_pet(STATES, PET_INFO):
    """Функция для кормления питомца."""
    if STATES['hunger'] == 0:
        print("Ваш питомец уже сыт!")
    else:
        STATES['hunger'] -= 20
        STATES['joy'] += 10
        print(f"{PET_INFO['name']} был накормлен.")


def play_with_pet(STATES, PET_INFO):
    """Функция для игры с питомцем."""
    if STATES['energy'] <= 30:
        print("У вашего питомца мало энергии, ему нужно отдохнуть.")
    else:
        STATES['joy'] += 15
        STATES['energy'] -= 25
        print(f"{PET_INFO['name']} получил удовольствие от игры.")


def sleep_pet(STATES, PET_INFO):
    """Функция для сна питомца."""
    if STATES['energy'] >= 90:
        print("Ваш питомец еще полон сил и не хочет спать.")
    else:
        STATES['energy'] += 50
        STATES['joy'] -= 10
        print(f"{PET_INFO['name']} поспал и восстановил силы.")


def show_pet_status(STATES, PET_INFO):
    """Функция для отображения состояния питомца."""
    print(f"Состояние {PET_INFO['name']}:")
    print(f"\tГолод: {STATES['hunger']}")
    print(f"\tРадость: {STATES['joy']}")
    print(f"\tЭнергия: {STATES['energy']}")


def update_states(STATES):
    """Функция для изменения состояний со временем."""
    STATES['hunger'] += 10
    STATES['joy'] -= 5
    STATES['energy'] -= 10
    for key in STATES.keys():
        if STATES[key] > 100:
            STATES[key] = 100
        elif STATES[key] < 0:
            STATES[key] = 0
````

````python
def main():
    """Основная программа."""
    print("Добро пожаловать в мир виртуальных питомцев!")
    while True:
        print('Что хотите сделать?')
        print('1. Усыновить питомца')
        print('2. Покормить питомца')
        print('3. Поиграть с питомцем')
        print('4. Уложить питомца спать')
        print('5. Проверить состояние питомца')
        print('6. Закончить игру')
        command = input('Ваш выбор: ')

        if command == '1':
            name = input('Как назовете своего питомца? ')
            pet_type = input('Какой у вас питомец? ')
            age = int(input('Сколько лет вашему питомцу? '))
            adopt_pet(name, pet_type, age, PET_INFO)
        elif command == '2':
            feed_pet(STATES, PET_INFO)
        elif command == '3':
            play_with_pet(STATES, PET_INFO)
        elif command == '4':
            sleep_pet(STATES, PET_INFO)
        elif command == '5':
            show_pet_status(STATES, PET_INFO)
        elif command == '6':
            break
        else:
            print("Неверная команда. Попробуйте снова.")

        update_states(STATES)


main()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random


knight_info = {
    'name': '',
    'armor': '',
    'weapon': '',
    'damage_range': (10, 20),
    'health': 100
}

dragon_info = {
    'name': 'Драконий Повелитель',
    'health': 150,
    'damage_range': (15, 25)
}


def introduction():
    """Введение в игру."""
    print("Добро пожаловать в битву между Рыцарем и Драконом!\n")
    knight_info['name'] = input("Как зовут вашего рыцаря? ")
    knight_info['armor'] = input("Какие доспехи носит ваш рыцарь? ")
    knight_info['weapon'] = input("Какое оружие использует ваш рыцарь? ")
    print(f"\nОтлично! Ваш рыцарь {knight_info['name']} готов к бою.\n")


def battle():
    """Управление боем."""
    print(f"Перед вами стоит грозный враг - {dragon_info['name']}.\n")
    while knight_info['health'] > 0 and dragon_info['health'] > 0:
            knight_attack()


def knight_attack():
    """Атакует рыцарь."""
    damage = random.randint(*knight_info['damage_range'])
    dragon_info['health'] -= damage
    print(f"{knight_info['name']} нанёс {damage} урона {dragon_info['name']}. Здоровье дракона: {dragon_info['health']}\n")
    if dragon_info['health'] > 0:
        dragon_attack()


def dragon_attack():
    """Атакует дракон."""
    damage = random.randint(*dragon_info['damage_range'])
    knight_info['health'] -= damage
    print(f"{dragon_info['name']} нанёс {damage} урона {knight_info['name']}. Здоровье рыцаря: {knight_info['health']}\n")


def end_battle():
    """Завершение боя."""
    if knight_info['health'] <= 0:
        print(f"Прощай, храбрый {knight_info['name']}! Ты пал в битве с {dragon_info['name']}.")
    else:
        print(f"Поздравляю! {knight_info['name']} победил грозного {dragon_info['name']}. Мир спасён!")
````

````python
def main():
    """Главная функция."""
    introduction()
    battle()
    end_battle()


main()
````
